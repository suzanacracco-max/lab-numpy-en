If a value in d is larger than d_mean but smaller than d_max, assign 75 to the corresponding value in f.
If a value equals to d_mean, assign 50 to the corresponding value in f.
Assign 0 to the corresponding value(s) in f for d_min in d.
Assign 100 to the corresponding value(s) in f for d_max in d.
In the end, f should have only the following values: 0, 25, 50, 75, and 100.
Note: you don't have to use Numpy in this question.



```python
### [your code here]
# Assume d is already defined

# Calculate statistics
d_min = d.min()
d_max = d.max()
d_mean = d.mean()

# Create empty array with same shape as d
f = np.empty(d.shape)

# Populate f based on rules
f[d == d_min] = 0               # d_min -> 0
f[(d > d_min) & (d < d_mean)] = 25  # d_min < d < d_mean -> 25
f[d == d_mean] = 50             # d == d_mean -> 50
f[(d > d_mean) & (d < d_max)] = 75  # d_mean < d < d_max -> 75
f[d == d_max] = 100             # d_max -> 100

# Verify f contains only the allowed values
print("Unique values in f:", np.unique(f))

# Print results
print("\nArray d:\n", d)
print("\nLabeled array f:\n", f)
```

### 18. Bonus question: instead of using numbers (i.e. 0, 25, 50, 75, and 100), how to use string values 
("A", "B", "C", "D", and "E") to label the array elements? You are expecting the result to be:
array([[[ 'D',  'D',  'D',  'B',  'D'],
        [ 'D',  'D',  'B',  'B',  'B'],
        [ 'D',  'B',  'D',  'D',  'D']],

       [[ 'B',  'B',  'B',  'B',  'E'],
        [ 'D',  'D',  'D',  'D',  'D'],
        [ 'B',  'D',   'A',  'D', 'D']]])
Again, you don't need Numpy in this question.

```python
### [your code here]

# Original array d as a Python nested list
d = [
    [[1.85836099, 1.67064465, 1.62576044, 1.40243961, 1.88454931],
     [1.75354326, 1.69403643, 1.36729252, 1.61415071, 1.12104981],
     [1.72201435, 1.1862918 , 1.87078449, 1.7726778 , 1.88180042]],

    [[1.44747908, 1.31673383, 1.02000951, 1.52218947, 1.97066381],
     [1.79129243, 1.74983003, 1.96028037, 1.85166831, 1.65450881],
     [1.18068344, 1.9587381 , 1.00656599, 1.93402165, 1.73514584]]
]

# Compute min, max, mean manually
flat = [x for plane in d for row in plane for x in row]  # flatten 3D list
d_min = min(flat)
d_max = max(flat)
d_mean = sum(flat)/len(flat)

# Create empty f with same shape
f = []

for plane in d:
    f_plane = []
    for row in plane:
        f_row = []
        for value in row:
            if value == d_min:
                f_row.append('A')
            elif d_min < value < d_mean:
                f_row.append('B')
            elif value == d_mean:
                f_row.append('C')
            elif d_mean < value < d_max:
                f_row.append('D')
            elif value == d_max:
                f_row.append('E')
        f_plane.append(f_row)
    f.append(f_plane)

# Print the labeled array
for plane in f:
    print(plane)
```
